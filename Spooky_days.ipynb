# How many Friday 13th in a year?

Riddle by stoned Elisina.

You can probably sort it out with math... but the brute force way is more tempting, given the irregular pattern of month length.

```python
# Create month in a year
jan = range(1,32)
feb = range(1,29)
febbis = range(1,30)
mar = range(1,32)
apr = range(1,31)
may = range(1,32)
jun = range(1,31)
jul = range(1,32)
aug = range(1,32)
sep = range(1,31)
oct = range(1,32)
nov = range(1,31)
dec = range(1,32)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year_months = [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec]
bisyear_months = [jan, febbis, mar, apr, may, jun, jul, aug, sep, oct, nov, dec]

s = 0
for month in year_months:
    s += len(month)
print("Days in a year %i" % s)
s = 0
for month in bisyear_months:
    s += len(month)
print("Days in a year 'bisestile'%i" % s)

#       0       1      2     3      4      5      6
week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
print("Days in a week %i" % len(week))
```

```python
offset = 1 # From which day you want to start the year?

s = 0
spooky = 0
for im, month in enumerate(year_months):
    for day in month:
        print("year day %5i) year week) %5i %3i %s %s" % (s+1, 
                                                          int(s/7)+1, 
                                                          day, 
                                                          month_names[im], 
                                                          week[(s+offset)%7])
             )
        if day == 13 and ((s+offset)%7) == 4:
            print("Spoooooky day")
            spooky += 1
        s += 1
print("Total spooky", spooky)
```

```python
for offset in range(7):
    print("Year starting on ", week[offset], end=" ")
    s = 0
    spooky = 0
    for im, month in enumerate(year_months):
        for day in month:
            if day == 13 and ((s+offset)%7) == 4:
                spooky += 1
            s += 1
    print("Total spooky", spooky)
for offset in range(7):
    print("Year bisestile starting on ", week[offset], end=" ")
    s = 0
    spooky = 0
    for im, month in enumerate(bisyear_months):
        for day in month:
            if day == 13 and ((s+offset)%7) == 4:
                spooky += 1
            s += 1
    print("Total spooky", spooky)
```

So maximum three Friday 13th in a year.
You can compute the probability of have one, two, or three spooky days picking up a year at random, but the problem is actually deterministic. 
If you know the first day you know the answer, no need for probability to get involved.
